```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
fin = open("1.txt")
clusters = [[] for i in range(2)]

x1, y1 = 1, 6
x2, y2 = 1.5, 7
aa = y2 - y1
bb = -(x2 - x1)
cc = - x1 * y2 + x2 * y1

def func(x, y):
  return (aa * x + bb * y + cc > 0)

for i in range(100):
    s = fin.readline()
    x, y = list(map(float, s.split('\t')))
    if func(x, y):
        clusters[0].append([x, y])
    else:
        clusters[1].append([x, y])

colors = ['blue', 'red']

for i, cluster in enumerate(clusters):
    cluster_x, cluster_y = zip(*cluster)
    plt.scatter(cluster_x, cluster_y, color=colors[i], label=f'Cluster {i+1}', alpha=0.6)

tx = ty = dist1 = dist2 = 0
for clust in clusters:
    mn = float('inf')
    for j in clust:
        now = j
        d = 0
        for doght in clust:
          x, y = doght
          d += (((x - now[0])**2 + (y - now[1])**2)**0.5)
        if d < mn:
            mn = d
            tx = now[0]
            ty = now[1]
    dist1 += tx
    dist2 += ty

plt.legend()
plt.title('0')
plt.grid()
plt.show()

print((dist1/2.0*1000), (dist2/2.0*1000))
```
